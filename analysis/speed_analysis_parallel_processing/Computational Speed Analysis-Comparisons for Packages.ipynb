```python
import time
import pandas as pd

sample_reviews = pd.read_json("D:/git_repos/neoway-brand-sentiment/sample_data/reviews_1000.json", lines = True)
speed_dict = dict()
```

### Stanford NLP timing

```python
# Calculate speed for Stanford NLP
from pycorenlp import StanfordCoreNLP

nlp = StanfordCoreNLP('http://localhost:9000')
start_time = time.time()
count = 0
for reviews in sample_reviews["text"]:
    res = nlp.annotate(reviews,
                       properties={
                           'annotators': 'sentiment',
                           'outputFormat': 'json'
                       })
    count += 1
end_time = time.time()
speed_dict["stanford_nlp"] = end_time - start_time
```

### Vader timing

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
start_time = time.time()
count = 0
for reviews in sample_reviews["text"]:
    print(str(count/1000 * 100) + "%")
    analyser.polarity_scores(reviews)
    count += 1
end_time = time.time()
speed_dict["vader"] = end_time - start_time
```

### Textblob timing

```python
import textblob
from textblob import TextBlob

start_time = time.time()
count = 0
for reviews in sample_reviews["text"]:
    print(str(count/1000 * 100) + "%")
    text = TextBlob(reviews)
    text_sentiment = text.sentiment
    count += 1
end_time = time.time()
speed_dict["textblob"] = end_time - start_time
```

### NLTK timing

```python
import nltk
from nltk import sentiment
nltk.download('opinion_lexicon')
nltk.download('subjectivity')

start_time = time.time()
count = 0
for reviews in sample_reviews["text"]:
    print(str(count/1000 * 100) + "%")
    sentiment = nltk.sentiment.util.demo_liu_hu_lexicon(reviews)
    count += 1
end_time = time.time()
speed_dict["nltk"] = end_time - start_time
```

### Computational speed comparison of methods for 1000 reviews

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.bar(speed_dict.keys(), speed_dict.values())
ax.set_yscale('log')
ax.set_title("Runtime of different methods")
ax.set_xlabel("Methods")
ax.set_ylabel("Seconds")
plt.show()
```

```python
from spellchecker import SpellChecker

spell = SpellChecker()

# find those words that may be misspelled
misspelled_wordcount = 0

def fix_misspellings(review):
    global spell
    
    misspelled = spell.unknown(review)
    
    for i in range(len(review)):
        if review[i] in misspelled:
            review[i] = spell.correction(review[i])
    
    return review

start_time = time.time()
sample_reviews['text_split'] = sample_reviews['text'].apply(lambda x: x.split())
sample_reviews['text_split'] = sample_reviews['text_split'].apply(lambda x: fix_misspellings(x))
end_time = time.time()
print(str(end_time - start_time) + " seconds")
```

### Computational speed of Vader as size of reviews grow

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
vader_dict = dict()
analyser = SentimentIntensityAnalyzer()

for size in [1,10,100,1000]:
    start_time = time.time()
    for reviews in sample_reviews["text"].sample(n=size):
        analyser.polarity_scores(reviews)
    end_time = time.time()
    vader_dict[size] = end_time - start_time
    
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(vader_dict.keys(), vader_dict.values())
ax.set_title("Runtime test on Vader")
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("Number of reviews")
ax.set_ylabel("Seconds")
plt.show()
```

### Computational speed of Stanford NLP as size of reviews

```python
from pycorenlp import StanfordCoreNLP
stanford_nlp_dict = dict()
nlp = StanfordCoreNLP('http://localhost:9000')


for size in [1, 10, 100, 1000]:
    print(size)
    start_time = time.time()
    for reviews in sample_reviews["text"].sample(n=size):
        res = nlp.annotate(reviews,
                           properties={
                               'annotators': 'sentiment',
                               'outputFormat': 'json'
                           })
    end_time = time.time()
    stanford_nlp_dict[size] = end_time - start_time
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(stanford_nlp_dict.keys(), stanford_nlp_dict.values())
ax.set_title("Runtime test on Stanford NLP")
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel("Number of reviews")
ax.set_ylabel("Seconds")
plt.show()
```
