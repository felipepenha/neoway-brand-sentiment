```python
import pandas as pd
import numpy as np
import re
import time
```

```python
brandlist = pd.read_csv('../analysis/wordnet_db/wordnet_food_beverages_list.csv')
```

```python
reviews = pd.read_json('../data/reviews_1500k.json', lines=True)
```

```python
reviews.head()
```

```python
reviews.text = reviews.text.str.lower()
brandlist.word = brandlist.word.str.lower()
```

```python
sample = reviews[750000:1000000].text
```

```python
%%time
train_data = []
for i in sample:
    brands = []
    for brand in brandlist.word:
        for m in re.finditer(r'\b' + brand + r'\b', i):
            brands.append((m.start(0), m.end(0), "PRODUCT"))
    train_data.append((i, brands))
```

Weird terms in brand list like 'delicious', 'round', 'green'

```python
result4 = pd.DataFrame(train_data, columns=['text', 'entities'])
```

```python
result4.to_csv('spacy_train4.csv')
```
